```python
xdata=[[300], [600], [1200]]

xdata_4core=[[300], [600]]
ydata_4core_CRS=[[12], [29]]
ydata_4core_noCRS=[[1], [4]]

ydata_8core_CRS=[[8], [17], [30]]
ydata_8core_noCRS=[[1], [2], [5]]

ydata_16core_CRS=[[10], [13], [21]]
ydata_16core_noCRS=[[1], [1], [4]]

xdata_36core=[[600], [1200], [2700]]

ydata_36core_CRS=[[9], [20], [58]]
ydata_36core_noCRS=[[1], [3], [9]]

xdata_system=[[300], [600], [1200], [2700]]
ydata_system_CRS=[[12], [17], [21], [58]]
ydata_system_noCRS=[[1], [2], [4], [9]]
```

```python
from sklearn.linear_model import LinearRegression

regr_4core_CRS = LinearRegression()
regr_4core_noCRS = LinearRegression()

regr_8core_CRS = LinearRegression()
regr_8core_noCRS = LinearRegression()

regr_16core_CRS = LinearRegression()
regr_16core_noCRS = LinearRegression()

regr_36core_CRS = LinearRegression()
regr_36core_noCRS = LinearRegression()

regr_system_CRS = LinearRegression()
regr_system_noCRS = LinearRegression()

regr_4core_CRS.fit(xdata_4core,ydata_4core_CRS)
regr_4core_noCRS.fit(xdata_4core,ydata_4core_noCRS)

regr_8core_CRS.fit(xdata,ydata_8core_CRS)
regr_8core_noCRS.fit(xdata,ydata_8core_noCRS)

regr_16core_CRS.fit(xdata,ydata_16core_CRS)
regr_16core_noCRS.fit(xdata,ydata_16core_noCRS)

regr_36core_CRS.fit(xdata_36core,ydata_36core_CRS)
regr_36core_noCRS.fit(xdata_36core,ydata_36core_noCRS)

regr_system_CRS.fit(xdata_system,ydata_system_CRS)
regr_system_noCRS.fit(xdata_system,ydata_system_noCRS)

```

```python
from matplotlib import pyplot as plt

plt.scatter(xdata_4core, ydata_4core_CRS)
plt.scatter(xdata, ydata_8core_CRS)
plt.scatter(xdata, ydata_16core_CRS)
plt.scatter(xdata_36core, ydata_36core_CRS)

plt.plot(xdata_4core, regr_4core_CRS.coef_*xdata_4core + regr_4core_CRS.intercept_, label='4core with CRS')
plt.plot(xdata, regr_8core_CRS.coef_*xdata + regr_8core_CRS.intercept_, label='8core with CRS')
plt.plot(xdata, regr_16core_CRS.coef_*xdata + regr_16core_CRS.intercept_, label='16core with CRS')
plt.plot(xdata_36core, regr_36core_CRS.coef_*xdata_36core + regr_36core_CRS.intercept_, label='36core with CRS')
#plt.plot(xdata_system, regr_system_CRS.coef_*xdata_system + regr_system_CRS.intercept_, label='system')

plt.legend()
plt.xlabel('Number of cells')
plt.ylabel('Average convergence time (sec)')
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.scatter(xdata_4core, ydata_4core_noCRS)
plt.scatter(xdata, ydata_8core_noCRS)
plt.scatter(xdata, ydata_16core_noCRS)
plt.scatter(xdata_36core, ydata_36core_noCRS)

plt.plot(xdata_4core, regr_4core_noCRS.coef_*xdata_4core + regr_4core_noCRS.intercept_, label='4core with no CRS')
plt.plot(xdata, regr_8core_noCRS.coef_*xdata + regr_8core_noCRS.intercept_, label='8core with no CRS')
plt.plot(xdata, regr_16core_noCRS.coef_*xdata + regr_16core_noCRS.intercept_, label='16core with no CRS')
plt.plot(xdata_36core, regr_36core_noCRS.coef_*xdata_36core + regr_36core_noCRS.intercept_, label='36core with no CRS')
#plt.plot(xdata_system, regr_system_noCRS.coef_*xdata_system + regr_system_noCRS.intercept_, label='system no CRS')

plt.legend()
plt.xlabel('Number of cells')
plt.ylabel('Average convergence time (sec)')
plt.show()
```
