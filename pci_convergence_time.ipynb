```python
xdata1=[[300], [600], [1200]]
ydata1=[[8], [17], [30]]

xdata2=[[300], [600], [1200]]
ydata2=[[10], [13], [21]]

xdata3=[[600], [1200]]
ydata3=[[17], [21]]
```

```python
from sklearn.linear_model import LinearRegression
regr1 = LinearRegression()
regr1.fit(xdata1,ydata1)
print(regr1.coef_, regr1.intercept_)

regr2 = LinearRegression()
regr2.fit(xdata2,ydata2)
print(regr2.coef_, regr2.intercept_)

regr3 = LinearRegression()
regr3.fit(xdata3,ydata3)
print(regr3.coef_, regr3.intercept_)
```

```python
from matplotlib import pyplot as plt
plt.scatter(xdata1, ydata1)
plt.scatter(xdata2, ydata2)
plt.plot(xdata1, regr1.coef_*xdata1 + regr1.intercept_, label='8core')
plt.plot(xdata2, regr2.coef_*xdata2 + regr2.intercept_, label='16core')
plt.plot(xdata3, regr3.coef_*xdata3 + regr3.intercept_, label='system')
plt.legend()
plt.xlabel('Number of cells')
plt.ylabel('Average convergence time (sec)')
plt.show()
```
